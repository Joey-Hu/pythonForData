```python
# 字典列表转化为DataFrame
info_keys = ['description', 'group', 'id', 'manufacturer']
info = DataFrame(data, columns=info_keys)
info
```

```python
info.info()
```

```python
# 查看食物类别及数量
pd.value_counts(info.group)
```

```python
# 将营养成分做成表
nutrients = []
for rec in data:
    nutrient = DataFrame(rec['nutrients'])
    nutrient['id'] = rec['id']
    nutrients.append(nutrient)

nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients.info()
```

```python
# 丢弃重复项
nutrients.duplicated().sum()
nutrients = nutrients.drop_duplicates()
```

```python
# 由于两个df中都有“group”和“description”，为了区分，对其重命名：
col_mapping  = {
    'description': 'food',
    'group':'fgroup'}
```

```python
info = info.rename(columns=col_mapping, copy=False)
info.info()
```

```python
col_mapping = {
    'description':'nutrient',
    'group':'nutgroup'}
```

```python
nutrients = nutrients.rename(columns=col_mapping, copy=False)
nutrients.info()
```

```python
# 合并nutrients和info
nutrients[:10]
```

```python
info[:10]
```

```python
ndata = info.merge(nutrients, on='id', how='outer')
nutrients.count()
# ndata[:55]
# ndata.info()
# ndata[ndata['id']==1008].count()
```

```python
# 图形化
result = ndata.groupby(['nutrient','fgroup'])['value'].quantile(0.5)
# result['Zinc, Zn'].order().plot(kind='barh')
Series.sort_values(result['Zinc, Zn']).plot(kind='barh')
```
